## Feature Engineering

```python
df = df.drop("name", axis=1)
```

```python
df = pd.get_dummies(df)
```

```python
df.head()
```

## Modelling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
df_scaled = scaler.fit_transform(df)
```

Using KMeans

```python
from sklearn.cluster import KMeans
```

```python
WCSS = []
for k in range(1,20):
    model = KMeans(n_clusters=k, n_init='auto')
    model.fit(df_scaled)
    WCSS.append(model.inertia_)
```

```python
plt.plot(range(1,20), WCSS)
plt.title("WCCS of clusters")
plt.xlabel("No of clusters")
plt.ylabel("WCCS")
```

```python
from sklearn.metrics import silhouette_score
```

```python
Silhouette_scores = []
for k in range(2,20):
    model = KMeans(n_clusters=k, n_init='auto')
    label = model.fit_predict(df_scaled)
    Silhouette_scores.append(silhouette_score(df_scaled, label))
```

```python
plt.plot(range(2,20), Silhouette_scores)
plt.title("Silhouette Score of Clusters")
plt.xlabel("No of clusters")
plt.ylabel("Silhouette Score")
```

```python
model = KMeans(n_clusters=4, n_init='auto')
```

```python
label = model.fit_predict(df_scaled)
```

```python
sns.scatterplot(x="mpg", y="horsepower", data=df, hue=label)
```

```python
sns.scatterplot(x="mpg", y="displacement", data=df, hue=label)
```

## Hierarchical Clustering

```python
from sklearn.cluster import AgglomerativeClustering
```

### Clustermap for hierarchical structure

```python
plt.figure(figsize=(15,5))
sns.clustermap(df_scaled, row_cluster=False)
```

```python
plt.figure(figsize=(15,5))
sns.clustermap(df_scaled, col_cluster=False)
```

Exploring ranges of cluster values

```python
from sklearn.metrics import silhouette_score
```

```python
Silhouette_scores = []
for n in range(2,10):
    model = AgglomerativeClustering(n_clusters=n)
    label = model.fit_predict(df_scaled)
    Silhouette_scores.append(silhouette_score(df_scaled, label))
```

```python
plt.plot(range(2,10), Silhouette_scores)
plt.title("Silhouette Score of Clusters")
plt.xlabel("No of clusters")
plt.ylabel("Silhouette Score")
```

Best number of clusters is 4

```python
model = AgglomerativeClustering(n_clusters=4)
```

```python
label = model.fit_predict(df_scaled)
```
